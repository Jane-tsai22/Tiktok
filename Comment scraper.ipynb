```python
import os
import glob
import pandas as pd
import requests
import time


try:
    from google.colab import userdata
    RAPIDAPI_KEY = userdata.get('RapidAPI')

except ImportError:
    from dotenv import load_dotenv
    load_dotenv(override=True)
    RAPIDAPI_KEY = os.getenv("RAPIDAPI_KEY")
  


API_HOST = "tiktok-video-feature-summary.p.rapidapi.com"
API_URL = "https://tiktok-video-feature-summary.p.rapidapi.com/comment/list"


def fetch_comments(video_url):
    headers = {
        "Content-Type": "application/json",
        "x-rapidapi-host": API_HOST,
        "x-rapidapi-key": RAPIDAPI_KEY
    }
    
    all_comments = []
    cursor = "0"
    has_more = True
    page = 1
    
    while has_more:

        querystring = {"url": video_url, "count": "30", "cursor": str(cursor)}
        
        try:
            response = requests.get(API_URL, headers=headers, params=querystring)
            
            if response.status_code != 200:
                print(f" API request error: {response.status_code})")
                break
                
            data = response.json()
            
  
            comments_list = data.get('comments') or data.get('data', {}).get('comments')
            
            if comments_list:
                for c in comments_list:
                    all_comments.append({
                        'video_url': video_url,
                        'comment_id': c.get('cid', ''),
                        'user_nickname': c.get('user', {}).get('nickname', ''),
                        'comment_text': c.get('text', ''),
                        'like_count': c.get('digg_count', 0),
                        'reply_comment_total': c.get('reply_comment_total', 0)
                    })
                
                has_more_flag = data.get('has_more') if 'has_more' in data else data.get('data', {}).get('has_more', 0)
                has_more = (has_more_flag == 1 or has_more_flag is True)
                
                new_cursor = data.get('cursor') if 'cursor' in data else data.get('data', {}).get('cursor', cursor)
                cursor = new_cursor
                
                print(f"Page {page} is scraped, total comments: {len(all_comments)} ")
                page += 1
            else:
                print(" All comments are scraped.")
                break 
                
        
            time.sleep(1.5)
                
        except Exception as e:
            print(f" Error: {e}")
            print("  Error reason:", data)
            break
            
    return all_comments


def main():
    input_folder = "/Users/a0979/Desktop/tiktok/Data/7052019-09062024/Split_Data/with_account/Scraped_Results"
    search_pattern = os.path.join(input_folder, "*_official.xlsx")
    all_files = glob.glob(search_pattern)
    
    if not all_files:
        print(f"Can't find the file：{input_folder}")
        return

    print(f"There are {len(all_files)} files.")

    for file_path in all_files:
        file_name = os.path.basename(file_path)
        if file_name.startswith("~$"): continue
            
        print(f"\n" + "="*50)
        print(f"Working on：【{file_name}】")
        
        output_filename = file_path.replace(".xlsx", "_comments.xlsx")
        
        try:
            df = pd.read_excel(file_path)
            party_comments = []
            

            if os.path.exists(output_filename):
                existing_comments_df = pd.read_excel(output_filename)
                processed_urls = existing_comments_df['video_url'].unique().tolist()
                party_comments = existing_comments_df.to_dict('records')
                print(f" Existed scraped comments :{len(processed_urls)}, skip them. ")
            else:
                processed_urls = []

            for index, row in df.iterrows():
                video_url = row.get('real_video_url', row.get('url')) 
                if not video_url or pd.isna(video_url): continue
                
                if video_url in processed_urls:
                    continue
                    
                print(f"  [{index+1}/{len(df)}] scraping video: {video_url}")
                
                comments = fetch_comments(video_url)
                
       
                if not comments:
                    comments = [{
                        'video_url': video_url,
                        'comment_id': 'NONE',
                        'user_nickname': 'System_Auto_Mark',
                        'comment_text': 'NO_COMMENTS_FOUND',
                        'like_count': 0,
                        'reply_comment_total': 0
                    }]
                
                party_comments.extend(comments)
                
              
                comments_df = pd.DataFrame(party_comments)
                comments_df.to_excel(output_filename, index=False)
                
        except Exception as e:
            print(f"File: {file_name} has error: {e}")

    print("\n✅ Completed!")

if __name__ == "__main__":
    main()
```
