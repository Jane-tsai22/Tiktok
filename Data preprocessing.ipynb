```python
import pandas as pd
```

```python
df=pd.read_excel("/Users/a0979/Desktop/tiktok/politici_Tiktok_27052019-09062024.xlsx")
```

```python
len(df)
```

```python
df = df.dropna(subset=['full', 'party'])

df['Tiktok'] = df['Tiktok'].astype(str)


df_with_account = df[df['Tiktok'].str.startswith('http')].copy()


df_no_account = df[~df['Tiktok'].str.startswith('http')].copy()


df_no_account['search_keyword'] = df_no_account['full'] + " " + df_no_account['party']


print(f"Politicians with an account：{len(df_with_account)}")
print(f"Politicians without an account：{len(df_no_account)}")


df_with_account.to_excel("politici_with_tiktok_account.xlsx", index=False)
df_no_account.to_excel("politici_without_tiktok_account.xlsx", index=False)
```

```python
df_with_account = pd.read_excel("politici_with_tiktok_account.xlsx")
df_no_account = pd.read_excel("politici_without_tiktok_account.xlsx")


for party_name, group_df in df_with_account.groupby('party'):

    safe_party_name = party_name.replace("&", "and") 
    filename = f"{safe_party_name}_with_account.xlsx"
    group_df.to_excel(filename, index=False)
    print(f"saved as : {filename} with {len(group_df)} data")
```

```python
for party_name, group_df in df_no_account.groupby('party'):
    safe_party_name = party_name.replace("&", "and")
    filename = f"{safe_party_name}_no_account.xlsx"
    group_df.to_excel(filename, index=False)
    print(f"saved as : {filename} with {len(group_df)} data")
```
